```python
import pandas as pd
import networkx as nx

nodes = pd.read_csv("nodes.csv")
edges = pd.read_csv("edges.csv")

# Step 1: Keep only edges where BOTH Source and Target 
# look like real HuggingFace model IDs (must contain '/')
valid_model_pattern = r'^[a-zA-Z0-9_\-\.]+/[a-zA-Z0-9_\-\.]+$'

clean_edges = edges[
    edges['Source'].str.match(valid_model_pattern, na=False) &
    edges['Target'].str.match(valid_model_pattern, na=False)
]

# Step 2: Separate by confidence for your research
edges_high   = clean_edges[clean_edges['Confidence'] == 1.0]  # explicit metadata
edges_strong = clean_edges[clean_edges['Confidence'] >= 0.9]  # + tag heuristics
edges_all    = clean_edges                                      # everything valid

print("High confidence only (explicit):", len(edges_high))
print("Strong confidence (≥0.9):", len(edges_strong))
print("All valid edges:", len(edges_all))

print("\nTransformation breakdown (strong):")
print(edges_strong['Transformation'].value_counts())
```
